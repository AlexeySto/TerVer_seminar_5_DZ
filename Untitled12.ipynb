1.	Когда используется критерий Стьюдента, а когда Z –критерий?

Ответ:

1. Критерий Стьюдента (t-критерий) обычно применяется, когда дисперсия генеральной совокупности неизвестна и рассчитывается на основе выборочных данных. Этот критерий более подходит для небольших выборок (обычно менее 30 наблюдений) и для случаев, когда дисперсия неизвестна.

2. Z-критерий, с другой стороны, применяется, когда дисперсия генеральной совокупности известна.


2.	Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

Решение:

Для проверки данной гипотезы будем использовать z-критерий.

Сформулируем нулевую и альтернативную гипотезы:

H_0: μ=17 мм (средний диаметр шарика равен 17 мм)

H_1: μ>17 мм (средний диаметр шарика больше 17 мм)

Z=  (X ̅- μ)/(σ/√n)=  (17,5-17)/(√4  / √100)  =  0,5/0,2=2,5

α=0,05=>Z_кр=1,645

Так как Z > Zкрит, мы отвергаем нулевую гипотезу в пользу альтернативной гипотезы.

Ответ: отвергаем нулевую гипотезу в пользу альтернативной.

4. Есть ли статистически значимые различия в росте дочерей? Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
import numpy as np
from scipy.stats import ttest_ind

def test_hypothesis(mothers_heights, daughters_heights):
    # Проведем двусторонний тест Стьюдента для независимых выборок
    t_stat, p_val = ttest_ind(mothers_heights, daughters_heights)

    alpha = 0.05  # Уровень значимости

    if p_val < alpha:
        print("Отклоняем нулевую гипотезу, есть статистически значимые различия в росте дочерей.")
    else:
        print("Нет статистически значимых различий в росте дочерей.")

# Рост матерей и дочерей
mothers_heights = [172, 177, 158, 170, 178, 175, 164, 160, 169, 165]
daughters_heights = [173, 175, 162, 174, 175, 168, 155, 170, 160]

test_hypothesis(mothers_heights, daughters_heights)
```
